```python
import numpy as np
from numpy.random import randn

#統計に関するモジュール
from scipy import stats

import seaborn as sns

%matplotlib inline
```

### boxplot 等

```python
data1 = randn(100)
data2 = randn(100)+2
```

```python
sns.distplot(data1)
sns.distplot(data2)
```

```python
sns.boxplot(data= [data1,data2])
# 箱ひげ図
```

```python
sns.boxplot(data=[data1,data2], whis= np.inf)
```

```python
sns.boxplot(data=[data1,data2], orient='h')

# orient は報告を決定する
```

```python
data1 = stats.norm(0,5).rvs(100)
# 平均が0標準偏差が5
```

```python
data2 = np.concatenate([stats.gamma(5).rvs(50)-1, -1*stats.gamma(5).rvs(50)])
# 2つのデータを繋げる
# gamma関数から50個データを取ってくるものと、-1をかけたものを繋げる
```

```python
sns.boxplot(data=[data1,data2])
```

```python
sns.violinplot(data = [data1,data2])
#　バイオリンプロット
# 　カーネル密度推定を使っている
```

```python
sns.violinplot(data = [data1,data2], bw=0.01)
# バンド幅を変えることができる
```

```python
sns.violinplot(data = [data1,data2], bw=0.01)
# バンド幅を変えることができる
```
