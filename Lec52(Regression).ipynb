```python
import numpy as np
from numpy.random import randn

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
# 食事の際などのtipのデータ
```

```python
tips
# dataframe
```

```python
sns.lmplot('total_bill','tip',tips)
# plotとRegression
# x軸にtotal_bill
# y軸にtip

# 色がついているところが信頼区間
```

```python
sns.lmplot('total_bill','tip',
           tips,scatter_kws={'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill',
           'tip',
           tips,
           order=4,
           scatter_kws={'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill',
           'tip',
           tips,
           scatter_kws={'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'},
          fit_reg=False)
```

```python
tips['tip_pect'] = 100*(tips['tip']/tips['total_bill'])

# 総額のナンパーセントが支払われたか
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)

# 離散的なプロットも可能
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=0.2)

# jitter プロットを横に広げて見やすくする
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('size','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True)
# 滑らかな回帰直線を描画する
# 局所的にデータをみている
```

```python
sns.regplot('total_bill','tip_pect',tips)

# 回帰直線だけでいいならregplot
```

```python
fig , (axis1,axis2) = plt.subplots(1,2,sharey= True)
sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(y='tip_pect',x='size',data= tips.sort_values('size'),ax=axis2)

# dataframe sort_valuesとsortの違いについて調べる
```
