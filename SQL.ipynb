```python
# sql文を入れるとデータフレームが作成される関数を用意した
def sql_to_df(sql_query):
    df = pd.read_sql(sql_query, con)
    return df
```

```python
query = '''select first_name , last_name 
                from customer'''
```

```python
sql_to_df(query)
```

```python
# *を使うと列名が全部返ってくる
query = 'select * from customer'
sql_to_df(query)
```

```python
sql_to_df('select * from city')
```

```python
sql_to_df('select country_id from city')

# city　テーブルからカラムcounty_idの列を取得
```

```python
sql_to_df('select distinct(country_id) from city')
#複数個の情報をまとめて取得する　ダブりをなくすイメージ
```

```python
sql_to_df('select * from customer where store_id = 1').head()
```

```python
sql_to_df('''select * from customer where first_name = 'MARY' ''').head()
```

```python
sql_to_df('''select * from film where release_year = 2006 and rating='R' ''').head()

#複数条件の指定も可能
```

```python
sql_to_df('''select * from film 
                      where release_year = 2006 or rating='R' ''').head()

#複数条件の指定も可能 or 指定でどちらかの条件の指定
```
