```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
dframe_wine = pd.read_csv('winequality-red.csv',sep=';')
```

```python
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
dframe_wine.sort_values('alcohol',ascending=False,inplace=True)

# ascendingは数字の並び順 Falseは降順、Trueは昇順
# inplaceはデータを書き換えるかどうか
```

```python
dframe_wine = dframe_wine.groupby('quality').apply(ranker)

# 元のワインのデータフレームを['quality']で分割して、その分割したデータフレームをranker()に入れて、アルコール度数のランキングのデータをマージして、データフレームを返している
```
